```python
import numpy as np
import pandas as pd
```

```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc("font", size = 18) 
```

```python
phones = pd.read_csv(r"C:/Users/Lord/Desktop/DataSets/1.8_phones.csv")

phones.head(15)
```

```python
fig = plt.figure(figsize=[10,10])

markers = {"Android": "o", "iOS": "*"}
for d1, d2, os in zip(phones.year, phones.price, phones.os):
    plt.scatter(d1,d2, s=250, marker=markers[os])
    
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
X = phones[["year","price"]]

clust = KMeans(n_clusters=2).fit(X)
```

```python
[c1, c2] = clust.cluster_centers_
```

```python
clust.cluster_centers_
```

```python
fig = plt.figure(figsize=[10,10])

markers = {"Android": "o", "iOS": "*"}
for d1, d2, os in zip(phones.year, phones.price, phones.os):
    plt.scatter(d1,d2, s=250, marker=markers[os])

plt.scatter(c1[0], c1[1], s=250, marker="x", c = "black")

plt.scatter(c2[0],c2[1], s=250, marker="x", c = "black")

plt.show()
```
